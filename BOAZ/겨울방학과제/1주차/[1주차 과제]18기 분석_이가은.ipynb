### 6. grade별로 평균 가격을 보여주세요.

```python
train_6 = train.groupby('grade')['price'].agg([('grade_mean','mean')])
train_6
```

```python
pd.merge(train, train_6, left_on='grade', right_on='grade', how='inner')
```
